```python
import os
from dotenv import load_dotenv
load_dotenv()
groq_api_key = os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq
model = ChatGroq(model="meta-llama/llama-4-scout-17b-16e-instruct",groq_api_key=groq_api_key)
model
```

```python
from langchain_core.messages import HumanMessage
model.invoke([HumanMessage(content = "Hi, my name is Manoj and I am a principal Technical consultant")])
```

```python
from langchain_core.messages import AIMessage
model.invoke(
    [HumanMessage(content = "Hi, my name is Manoj and I am a principal Technical consultant"),
    AIMessage(content = "Hello Manoj! It's nice to meet you. As a Principal Technical Consultant, you must have a wealth of experience and expertise in your field. What area of technology do you specialize in, if you don't mind me asking?"),
    HumanMessage(content = "what is my name and what do I do?"),
    ])
```

```python
### Message History
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

store = {}

def get_session_history(session_id:str)->BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return  store[session_id]

with_msg_history = RunnableWithMessageHistory(model,get_session_history)

config = {"configurable": {"session_id":"chat1"}}

response = with_msg_history.invoke([HumanMessage(content = "Hi, my name is Manoj and I am a principal Technical consultant")],
config = config
)
response.content
```

```python
with_msg_history.invoke([HumanMessage(content = "what is my name?")],
config = config
)
```

```python
config2 = {"configurable": {"session_id":"chat2"}}
with_msg_history.invoke([HumanMessage(content = "what is my name?")],
config = config2
)
store
```

Prompt Templates

```python
from langchain_core.prompts import ChatPromptTemplate,MessagesPlaceholder

prompt = ChatPromptTemplate.from_messages([
                ("system", "You are a helpful AI bot.Answer all questions to your best of ability"),
                MessagesPlaceholder(variable_name="messages")
                ])

chain = prompt|model
chain.invoke({"messages": [HumanMessage(content = 'Hi My Name is Manoj')]})
```

```python
with_msg_history = RunnableWithMessageHistory(model, get_session_history)
config = {"configurable": {"session_id": "chat3"}}
with_msg_history.invoke( [HumanMessage(content = 'Hi My Name is Manoj')], config)
```

```python
## Add More Complexity
prompt_hin = ChatPromptTemplate.from_messages([
                ("system", "You are a helpful AI bot.Answer all questions to your best of ability in {language}."),
                MessagesPlaceholder(variable_name="messages")
                ])
chain_hin = prompt_hin|model
response_hin = chain_hin.invoke({"messages": [HumanMessage(content = 'Hi My Name is Manoj')], "language": "Telugu"})
response_hin.content
```

```python
with_msg_history = RunnableWithMessageHistory(chain_hin, get_session_history, input_messages_key="messages")
config = {"configurable": {"session_id":"chat5"}}
response = with_msg_history.invoke(
    {"messages": [HumanMessage(content = 'Hi My Name is Manoj')], "language": "Telugu"},
    config = config
    )
response.content
```

Managing the Conversation History

```python
from langchain_core.messages import SystemMessage,trim_messages
trimmer = trim_messages(
    max_tokens=70,
    strategy="last",
    token_counter=model,
    include_system=True,
    allow_partial=False,
    start_on="human"
)

messages = [
    SystemMessage(content="you're a good assistant"),
    HumanMessage(content="hi! I'm bob"),
    AIMessage(content="hi!"),
    HumanMessage(content="I like vanilla ice cream"),
    AIMessage(content="nice"),
    HumanMessage(content="whats 2 + 2"),
    AIMessage(content="4"),
    HumanMessage(content="thanks"),
    AIMessage(content="no problem!"),
    HumanMessage(content="having fun?"),
    AIMessage(content="yes!"),
]
trimmer_response = trimmer.invoke(messages)
trimmer_response
```

```python
from operator import itemgetter
from langchain_core.runnables import RunnablePassthrough

chain = (
    RunnablePassthrough.assign(messages= itemgetter("messages")|trimmer)
    | prompt_hin|model
)
chain.invoke({"messages":messages + [HumanMessage(content= "what ice cream do I like")],
"language":"English"
}
)
```
