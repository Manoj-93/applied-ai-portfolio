```python
from langchain_chroma import Chroma
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader = PyPDFLoader('attention.pdf')
pdf_doc = loader.load()
pdf_doc
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap =20)
splits = text_splitter.split_documents(pdf_doc)
```

```python
ollama_embedding =  ( OllamaEmbeddings(model = "embeddinggemma")) 
vectordb = Chroma.from_documents(documents = splits, embedding=ollama_embedding )
vectordb
```

```python
query_from_vec_db = "In a self-attention layer all of the keys, values and queries come from the same place"
docs = vectordb.similarity_search(query_from_vec_db)
docs
docs[0].page_content
```
