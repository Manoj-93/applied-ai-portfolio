How to recursively split text by characters

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
final_text_doc = text_splitter.split_documents(text_documents)
final_text_doc
print(final_text_doc[0])
print(final_text_doc[1])
print(final_text_doc[2])
```

HTML TEXT Splitter


```python
from langchain_text_splitters import HTMLHeaderTextSplitter
html_sring = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>My First Web Page</title>
</head>
<body>
    <h1>Welcome to My Website</h1>
    <p>This is a simple paragraph of text.</p>
</body>
</html>"""

headers_to_split_on = [("h1", "header 1")]
html_splitter = HTMLHeaderTextSplitter(headers_to_split_on)
html_headr_splits = html_splitter.split_text(html_sring)
html_headr_splits
```

```python
from langchain_text_splitters import HTMLHeaderTextSplitter
url = "https://docs.langchain.com/oss/python/integrations/document_loaders/arxiv"
headers_to_split_on = [("h1", "header 1"),("h2", "header 2"),("h3", "header 3")]
html_splitter = HTMLHeaderTextSplitter(headers_to_split_on)
html_headr_splits = html_splitter.split_text_from_url(url)
html_headr_splits
```

##FAISS

```python
from langchain_community.vectorstores import FAISS


vec_db = FAISS.from_documents(final_pdf_doc,ollama_embedding )
vec_db
```

```python
query_from_vec_db = "In a self-attention layer all of the keys, values and queries come from the same place"
docs = vec_db.similarity_search(query_from_vec_db)
docs
docs[0].page_content
```

Retriever

```python
retriever = vec_db.as_retriever()
docs_retriever = retriever.invoke(query_from_vec_db)
docs_retriever[0].page_content
```

Similarity Search

```python
dosc_score = vec_db.similarity_search_with_score(query_from_vec_db)
dosc_score
```

```python
vec_db.save_local("faiss_index")
```

```python
new_df = FAISS.load_local("faiss_index",ollama_embedding, allow_dangerous_deserialization=True )
new_dba = new_df.similarity_search(query_from_vec_db)
new_dba
```
