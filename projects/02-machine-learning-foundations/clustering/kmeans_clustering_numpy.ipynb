```python
import numpy as np

X = np.array([
    [1, 2],
    [1, 4],
    [1, 0],
    [10, 2],
    [10, 4],
    [10, 0]
])

K = 3


def k_means(X, K, max_iters=100):
    C_old = X[:K].copy()

    for _ in range(max_iters):
        C_new = find_clusters(C_old, X, K)

        if np.array_equal(C_old, C_new):
            break

        C_old = C_new

    return C_old


def find_clusters(C, X, K):
    new_centroids = []

    X_exp = X[:, None, :]
    C_exp = C[None, :, :]

    dist_matrix = np.sum((X_exp - C_exp) ** 2, axis=2)
    cluster_assignment = np.argmin(dist_matrix, axis=1)

    for i in range(K):
        mask = cluster_assignment == i
        cluster_points = X[mask]

        if len(cluster_points) == 0:
            centroid = C[i]
        else:
            centroid = np.mean(cluster_points, axis=0)

        new_centroids.append(centroid)

    return np.array(new_centroids)


final_centroids = k_means(X, K)
print("Final centroids:\n", final_centroids)





























```
