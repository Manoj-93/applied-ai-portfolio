Building area should be a float, Data column needs to be updated to date type from object

```python
# changing the data type of Date column

#data['Date'] = pd.to_datetime(data['Date'])
#data["Date"].unique()
data['Date'] = pd.to_datetime(data['Date'], format="%d-%m-%Y")
```

```python
#why BuildingArea column has object data type
data['BuildingArea'].unique()
```

```python

#replace inf, missing to NAN
data["BuildingArea"] = data["BuildingArea"].replace(["missing" , "inf"],np.nan)
#data["BuildingArea"].apply(type).value_counts()
data["BuildingArea"].unique()
```

```python
#convert BuildingArea to Float
data["BuildingArea"] =  data["BuildingArea"].astype(float)
data["BuildingArea"].apply(type).value_counts()
```
