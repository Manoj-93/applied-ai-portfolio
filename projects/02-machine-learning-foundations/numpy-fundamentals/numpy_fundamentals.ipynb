```python
import numpy as np

points = np.array([
    [1,2,3],
    [1,1,1],
    [3,2,1]
])

print("shape:", points.shape)

print("row 0:", points[0])
print("column 0:", points[:,0])
print(points[:,1])
print("mean axis 0:", np.mean(points, axis=0))
print("mean axis 1:", np.mean(points, axis=1))
```

```python
import numpy as np

points = np.array([
    [1,2,3],
    [1,1,1],
    [1,1,1],
    [3,2,3],
    [3,2,1]
])

centroids = np.array([
    [1,2,3],
    [1,1,1],
    [3,2,1]
])

print(points[:, np.newaxis, :])
diff = points[:, np.newaxis, :] - centroids
print("diff shape:", diff.shape)

dist = np.sqrt(np.sum(diff**2, axis=2))
print("distance matrix:\n", dist)

assignments = np.argmin(dist, axis=1)
print("assignments:", assignments)
```
