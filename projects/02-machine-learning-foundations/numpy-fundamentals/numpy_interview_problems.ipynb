```python
import numpy as np



items = np.array([
    [1. , 1.2],   # v1
    [0.8, 1. ],   # v2
    [8.8, 9. ],   # v3
    [1.2, 0.9],   # v4
    [1.2, 0.9]    # v5
])

k = 2
ran_vector = np.random.choice(len(items),k)
initial_centroids = items[ran_vector]
print(initial_centroids)
for i in range(len(initial_centroids)):
  for j in range(len(items)):
    dist = np.sum((items[j] - initial_centroids[i])**2)
    print(dist)


```
