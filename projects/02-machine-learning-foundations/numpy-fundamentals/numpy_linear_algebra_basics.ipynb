```python
import numpy as np


U = np.array([0.4, 0.6])
V_pos = np.array([0.8, 0.3])
V_neg1 = np.array([-0.2, 0.9])
V_neg2 = np.array([0.5, -0.7])

alpha = 0.1
lambda_reg = 0.01
steps = 20

def softmax(x):
  ex = np.exp(x - np.max(x))
  return ex/sum(ex)

for s in range(steps):
  score = [np.dot(U,V_pos),np.dot(U,V_neg1), np.dot(U,V_neg2)]
  scores = softmax(score)
  P_pos = scores[0]
  P_neg1 = scores[1]
  P_neg2 = scores[2]
  d_scores = [P_pos - 1,P_neg1, P_neg2]
  dU = d_scores[0] * V_pos + d_scores[1] * V_neg1 + d_scores[2] * V_neg2
  print(dU)
  dV_pos = d_scores[0] * U
  dV_neg1 = d_scores[1] * U
  dV_neg2 = d_scores[2] * U
  U = U - alpha * (dU + lambda_reg * U)
  V_pos = V_pos - alpha * (dV_pos + lambda_reg * V_pos)
  V_neg1 = V_neg1 - alpha * (dV_pos + lambda_reg * V_neg1)
  V_neg2 = V_neg2 - alpha * (dV_pos + lambda_reg * V_neg2)

  print("2d2", U)







```
