```python
import numpy as np

# -----------------------------
# INITIAL USER & ITEM EMBEDDINGS
# -----------------------------
U = np.array([0.5, 0.5])          # User embedding
V_pos = np.array([0.6, 0.4])      # Positive item (user clicked)
V_neg = np.array([-0.3, 0.6])     # Negative item (user did not click)

alpha = 0.1       # learning rate
lambda_reg = 0.01 # small L2 regularization
steps = 20        # number of training iterations

def softmax(x):
    ex = np.exp(x - np.max(x))  # numerical stability trick
    return ex / ex.sum()

for step in range(steps):

    # -----------------------------
    # 1. Compute dot-product scores
    # -----------------------------
    score_pos = np.dot(U, V_pos)
    score_neg = np.dot(U, V_neg)

    scores = np.array([score_pos, score_neg])

    # -----------------------------
    # 2. Compute softmax probabilities
    # -----------------------------
    probs = softmax(scores)
    P_pos = probs[0]
    P_neg = probs[1]

    # -----------------------------
    # 3. Cross-entropy loss
    # Since positive is class 0:
    # Loss = -log(P_pos)
    # -----------------------------
    loss = -np.log(P_pos)

    # -----------------------------
    # 4. Gradient of loss w.r.t scores
    # (softmax + cross entropy derivative)
    # dL/dscore_pos = P_pos - 1
    # dL/dscore_neg = P_neg - 0  since neg is 0-label
    # -----------------------------
    d_scores = np.array([P_pos - 1, P_neg])

    # -----------------------------
    # 5. Gradient of loss w.r.t embeddings
    # dL/dU = sum(d_scores[i] * V[i])
    # dL/dV = d_scores[i] * U
    # -----------------------------
    dU = d_scores[0] * V_pos + d_scores[1] * V_neg
    print(dU)
    dV_pos = d_scores[0] * U
    dV_neg = d_scores[1] * U

    # -----------------------------
    # 6. Gradient descent update
    # -----------------------------
    U = U - alpha * (dU + lambda_reg * U)
    V_pos = V_pos - alpha * (dV_pos + lambda_reg * V_pos)
    V_neg = V_neg - alpha * (dV_neg + lambda_reg * V_neg)

    print(f"STEP {step+1}")
    print("score_pos =", score_pos, " score_neg =", score_neg)
    print("P(pos) =", P_pos)
    print("Loss =", loss)
    print("Updated U:", U)
    print("Updated V_pos:", V_pos)
    print("Updated V_neg:", V_neg)
    print("-" * 40)
```
