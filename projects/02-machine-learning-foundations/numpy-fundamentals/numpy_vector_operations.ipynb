```python
import numpy as np

# Initial embeddings
U = np.array([0.5, 0.3])   # user embedding
V = np.array([0.9, 0.1])   # item embedding

alpha = 0.1      # learning rate
lambda_reg = 0.01  # small L2 regularization
true_label = 1.0   # user liked the item

print("Initial U:", U)
print("Initial V:", V)
print("===================")

for step in range(20):  # 20 small updates

    # 1) Prediction (dot product)
    pred = np.dot(U, V)

    # 2) Error
    error = pred - true_label

    # Save old copies for stable updates
    U_old = U.copy()
    V_old = V.copy()

    # 3) Gradient descent update
    U = U - alpha * (error * V_old + lambda_reg * U_old)
    V = V - alpha * (error * U_old + lambda_reg * V_old)

    print(f"Step {step+1}")
    print("Prediction:", pred)
    print("Error:", error)
    print("Updated U:", U)
    print("Updated V:", V)
    print("-------------------")
```
