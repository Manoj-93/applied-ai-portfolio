```python
import csv
def expense_details():
    date = input("Enter date in YYYY-MM-DD format: ")
    category = input("category of the expense: ")
    amt_spent = float(input("Enter amount spent: "))
    description = input("Enter description ")
    expense = {
       "date": date,
       "category": category,
       "amt_spent": amt_spent,
       "description": description
     }
    return expense

def display_expenses(expenses):
    for i in range(len(expenses)):
        item = expenses[i]

        print("\nExpense", i + 1)

        for key, value in item.items():
            if key == "amt_spent" and value <= 0:
                print(key, "is empty or invalid")
            elif value == "":
                print(key, "is empty")
            else:
                print(key, ":", value)
def montly_budget():
    montly_budget = float(input("Enter the total amount they want to budget for the month"))
    return montly_budget
def total_budget_so_far(expenses):
    total_amt_so_far = 0;
    for i in range(len(expenses)):
        item = expenses[i]
        for key, value in item.items():
            if key == "amt_spent":
                total_amt_so_far = total_amt_so_far + value
    return total_amt_so_far
def save_expenses_to_csv(expenses):
    file = open("expenses.csv", "w", newline="")
    writer = csv.writer(file)

    writer.writerow(["date", "category", "amt_spent", "description"])

    for item in expenses:
        writer.writerow([
            item["date"],
            item["category"],
            item["amt_spent"],
            item["description"]
        ])

    file.close()
def load_expenses_from_csv():
    expenses = []

    try:
        file = open("expenses.csv", "r")
    except FileNotFoundError:
        return expenses
    reader = csv.reader(file)

    header = next(reader)  # skip header

    for row in reader:
        expense = {
            "date": row[0],
            "category": row[1],
            "amt_spent": float(row[2]),
            "description": row[3]
        }
        expenses.append(expense)

    file.close()
    return expenses

def interactive_menu():
    expense = load_expenses_from_csv()
    budget = 0.0

    while True:
        print("\n1: Add expense")
        print("2: View expenses")
        print("3: Track budget")
        print("4: Save expenses")
        print("5: Exit")

        menu_selected = int(input("Enter a number to choose an option: "))

        if menu_selected == 1:
            det_from_fun = expense_details()
            expense.append(det_from_fun)
            print("Expense added.")

        elif menu_selected == 2:
            display_expenses(expense)

        elif menu_selected == 3:
            if budget == 0:
                budget = montly_budget()

            total_amt_so_far = total_budget_so_far(expense)

            if total_amt_so_far > budget:
                print("You have exceeded your budget!")
            else:
                print("You have", budget - total_amt_so_far, "left for the month")

        elif menu_selected == 4:
            save_expenses_to_csv(expense)
            print("Expenses saved.")

        elif menu_selected == 5:
            save_expenses_to_csv(expense)
            print("Saved. Exiting now.")
            break

        else:
            print("Invalid option. Try again.")

interactive_menu()
```
