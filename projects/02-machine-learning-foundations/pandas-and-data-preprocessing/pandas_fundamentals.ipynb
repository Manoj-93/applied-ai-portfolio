```python
import pandas as pd

df = pd.DataFrame({
    "user_id": [1,2,3,4,5,6],
    "age": [25, 34, None, 45, 23, 36],
    "department": ["Sales", "Tech", "Tech", "HR", "Sales", "Tech"],
    "salary": [50000, 80000, 75000, None, 52000, 90000],
    "join_date": ["2020-01-15", "2019-03-10", "2021-06-01", "2018-07-23", "2022-02-20", "2020-11-11"],
    "left_company": [0,1,0,1,0,1]
})
df.loc[df["age" ] >30, 'user_id']
df.loc[(df["salary"] > 6000) & (df['department'] == "Tech"), "user_id" ]
df.loc[ (df["age"]>30) | (df["salary"] >100000)  , "user_id"]
```

```python
import pandas as pd

users = pd.DataFrame({
    "user_id": [1, 2, 3, 4],
    "department": ["IT", "HR", "IT", "Finance"],
    "salary": [90000, 60000, 110000, 75000]
})

performance = pd.DataFrame({
    "user_id": [1, 2, 3, 5],
    "performance_score": [8, 7, 9, 6]
})
df = pd.merge (users,performance , on = "user_id", how = 'outer'  )
df
df.loc[df['salary'].isna(), "user_id"]
```
