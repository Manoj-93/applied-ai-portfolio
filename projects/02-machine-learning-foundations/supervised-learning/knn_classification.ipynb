```python
import numpy as np
X_train = np.array([
[1, 2],
[2, 3],
[3, 3],
[6, 5],
[7, 7],
[8, 6]])
k = 3
label = [0,2,2,2,1,1]
Y = []
nearest_neighbours = []

x_test = [5, 5]
squared_dist = (X_train - x_test)**2
dist_matrix = np.sum(squared_dist, axis=1 )
print(dist_matrix)

for i in range(k):
  print("i is",i)
  smaller_value = dist_matrix[0]
  for index, dist_value in enumerate(dist_matrix):
    if index not in Y:
     if smaller_value >= dist_value:
        smaller_value =dist_value
        small_index = index
        print("smller_value ",smaller_value)
        print("small_index ", small_index)

  Y.append(small_index)
  print("y is",Y)
for index, val_final in enumerate(Y):
  while len(nearest_neighbours) <= val_final:
    nearest_neighbours.append(0)
  nearest_neighbours[label[val_final]] = nearest_neighbours[label[val_final]]+1
  print(nearest_neighbours)
print(np.argmax(nearest_neighbours))



















```
