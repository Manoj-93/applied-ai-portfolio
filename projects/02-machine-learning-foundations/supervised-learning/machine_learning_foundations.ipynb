Machine Learning Fundamentals тАФ Structured Notes
1. Data Science: The Core Question

Data science tries to answer the question:

Can we learn about the world using data?

The general process is:

Collect data

Find patterns in the data

Represent those patterns using mathematical models

So the goal is to extract knowledge from data.

Professional framing:

Data science uses statistical and computational techniques to extract meaningful patterns and insights from data.

2. Machine Learning: Automating Model Building

Machine learning is a subset of data science.

Instead of manually designing models, machine learning:

Takes data as input

Discovers patterns automatically

Produces a mathematical model

So the pipeline becomes:

Data тЖТ Pattern Learning тЖТ Mathematical Model тЖТ Predictions

Professional framing:

Machine learning automates the process of discovering patterns in data and building predictive models.

3. The Core Challenge: Signal vs Noise

Real-world data contains two components:

Component	Meaning
Signal	Real patterns that describe the world
Noise	Random variation or irrelevant data

Example:

Housing prices may depend on:

location

square footage

school district

But noise could include:

random market fluctuations

unusual sales

measurement errors

The challenge in ML is:

Identify the signal while ignoring the noise.

4. Training vs Testing Data

Machine learning models must generalize to new data.

So we split the dataset into:

Dataset	Purpose
Training Set	Used to learn patterns
Testing Set	Used to evaluate the model

Pipeline:

Training Data тЖТ Learn Model
Testing Data тЖТ Evaluate Model Performance

If the model performs well on unseen data, it has learned real patterns.

5. Overfitting vs Underfitting

Two common model failures occur during learning.

Overfitting

The model learns signal + noise.

Symptoms:

Very low training error

High testing error

Example:

A model memorizes the training data rather than learning general patterns.

Professional definition:

Overfitting occurs when a model captures noise in the training data and fails to generalize.

Underfitting

The model is too simple and fails to capture patterns.

Symptoms:

High training error

High testing error

Professional definition:

Underfitting occurs when the model lacks sufficient complexity to capture underlying patterns.

6. Model Complexity Tradeoff

There is a balance between model complexity and prediction error.

Typical relationship:

Low complexity тЖТ Underfitting
Optimal complexity тЖТ Best generalization
High complexity тЖТ Overfitting

Goal:

Find the model complexity that captures signal but ignores noise.

This is sometimes called the biasтАУvariance tradeoff.

2. Correlation

Correlation measures the strength and direction of a relationship between two variables.

The value of correlation always ranges between:

-1  тЙд  Correlation тЙд  +1
Interpretation

4. Pearson Correlation Coefficient

The most common correlation measure is PearsonтАЩs correlation coefficient.

Formula conceptually measures:

5. Important Concept: Correlation тЙа Causation

One of the most important concepts in data science.

Just because two variables are correlated does NOT mean one causes the other.

Example:

Ice cream sales тЖС
Shark attacks тЖС

Both increase during summer.

But:

Ice cream does NOT cause shark attacks.

The true cause is:

This is called a confounding variable.

8. Simple Linear Regression

The simplest regression model assumes a linear relationship.

Meaning:

Relationship follows a straight line

Mathematical equation:

y = b0 + b1x

Where:

y  = predicted output (sales)
x  = input variable (advertising spend)

b0 = intercept
b1 = slope
9. Understanding the Intercept (b0)

The intercept represents:

Predicted value of y when x = 0

Example:

Sales = 1.01 + 2.45 ├Ч Advertising

When advertising = 0:

Sales = 1.01

This is the baseline sales level.

10. Understanding the Slope (b1)

The slope represents:

Change in y for a 1 unit change in x

Example:

Sales = 1.01 + 2.45 ├Ч Advertising

Interpretation:

For every 1 unit increase in advertising,
sales increase by 2.45 units.

Slope tells us how strongly the input influences the output.

11. Interpreting Regression Coefficients

Example model from lecture:

sales = 1.01 + 2.45 ├Ч advertising

Interpretation:

Term	Meaning
1.01	baseline sales
2.45	impact of advertising

Meaning:

Advertising increases sales.

But this interpretation is valid only if model assumptions hold.

12. Best-Fit Line

In real datasets, data points do not lie perfectly on a line.

Example scatter plot:

Advertising vs Sales

Points are scattered.

The regression algorithm finds a line that best represents the relationship.

This is called the:

Best-fit line
13. Prediction Error

Since predictions are not perfect, we measure prediction error.

Formula:

Error = Actual Value тИТ Predicted Value

Example:

Actual sales = 14,000
Predicted sales = 13,040
Error = 960
14. How the Best-Fit Line is Chosen

There can be many possible lines.

The correct line is chosen using Least Squares.

Goal:

Minimize total prediction error.

More specifically:

Minimize sum of squared errors.

Mathematically:

╬г (Actual тИТ Predicted)┬▓

Squaring ensures:

Large errors are penalized heavily

This produces the optimal regression line.
