```python
import numpy as np

# 5 items, 4-dim embeddings
items = np.array([
    [1.0, 1.2, 5.0, 5.2],   # v1
    [0.8, 1.0, 4.8, 5.0],   # v2
    [9.0, 9.2, 0.5, 0.6],   # v3
    [8.8, 9.0, 0.4, 0.7],   # v4
    [1.2, 0.9, 4.9, 5.1]    # v5
])

k = 2  # number of IVF clusters (centroids)

# Pretend these came from your K-Means loop:
centroids = np.array([
    [0.999, 1.033, 4.9, 5.1],   # centroid for cluster 0  (close to v1, v2, v5)
    [8.9,   9.1,   0.45, 0.65]  # centroid for cluster 1  (close to v3, v4)
])

# Cluster assignments per item (one label per row in `items`)
# index:  0  1  2  3  4
cluster_assignments = np.array([0, 0, 1, 1, 0])

IVF = [[] for _ in range(k)]
for i in range(len(cluster_assignments)):
  IVF[cluster_assignments[i]].append(i)
  print(IVF)
u = np.array([1.0, 1.1, 5.1, 5.0])
dist_u = np.zeros(k)
for cent in range(k):
  print('centroids[cent]',centroids[cent])
  dist_u[cent] = np.sum((centroids[cent] - u) ** 2)

print(dist_u)
chosen_cluster = np.argmin(dist_u)
print('chosen_cluster',chosen_cluster)
for ivf in range(len(IVF[chosen_cluster])):
 print(np.dot(u,items[IVF[chosen_cluster][ivf]]))
```
