```python
class LoanApplication:
    def __init__(self, name, income, monthly_debt, credit_score):
        self.name = name
        self.income = income
        self._monthly_debt = monthly_debt
        self.credit_score = credit_score

app1 = LoanApplication("Manoj", 90000, 20000, 740)
app2 = LoanApplication("Ravi", 45000, 2000, 620)

print(app1._monthly_debt)         # Manoj
print(app2._monthly_debt) # 620
```

```python
class LoanApplication1:
    def __init__(self, credit_score):
        self._credit_score = credit_score

    @property
    def credit_score(self):
        return self._credit_score

    @credit_score.setter
    def credit_score(self, value):
        if 300 <= value <= 850:
            self._credit_score = value
        else:
            raise ValueError("Invalid credit score")
```

```python
app1 = LoanApplication1(1700)
```

```python
from abc import ABC, abstractmethod

class LoanApplication:
    def __init__(self, name, credit_score):
        self.name = name
        self.credit_score = credit_score


class DecisionEngine(ABC):
    @abstractmethod
    def decide(self, application):
        pass


class RuleBasedEngine(DecisionEngine):
    def decide(self, application):
        if application.credit_score >= 700:
            return "APPROVE"
        return "REJECT"


class ConservativeEngine(DecisionEngine):
    def decide(self, application):
        if application.credit_score >= 750:
            return "APPROVE"
        return "REJECT"


app = LoanApplication("Manoj", 720)

engine1 = RuleBasedEngine()
engine2 = ConservativeEngine()

print(engine1.decide(app))   # APPROVE
print(engine2.decide(app))   # REJECT
```

```python
class LoanApplication:
    def __init__(self, name, credit_score):
        self.name = name
        self.credit_score = credit_score

    def done(self):
        return f"{self.name} (score={self.credit_score})"
```

```python
app = LoanApplication("Manoj", 740)
print(app.done)
```
